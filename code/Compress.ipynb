```python
from PIL import Image
import PIL
import os
import glob
import numpy as np
```

```python
#file_path = 'CS_Dataset/v_ApplyEyeMakeup_g15_c03'
#file_path = 'CS_Dataset/Frames/v_PlayingCello_g01_c07'
file_path = 'CS_Dataset/Frames/v_HammerThrow_g01_c05'
newsize = (512, 512)
for i in [10,20,30,40,50]:
    for image in np.sort(os.listdir(file_path)):
        # 1. Open the image
        img = Image.open('CS_Dataset/Frames/v_HammerThrow_g01_c05/'+image)
        img = img.resize(newsize)
        # 2. Compressing the image
        img.save(str(i)+"/compress_"+image,
                 optimize=True,
                 quality=i)
```
